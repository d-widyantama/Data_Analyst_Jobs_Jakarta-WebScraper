```python
import requests
from bs4 import BeautifulSoup
```

```python

# Define the URL

# Set a user-agent header to mimic a browser

# Make the initial request
def extract(page):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'
    }
    url = f"https://id.linkedin.com/jobs/search?keywords=Data%20Analyst&location=Area%20DKI%20Jakarta&geoId=90010101&trk=public_jobs_jobs-search-bar_search-submit&position=1&pageNum={page}"
    response = requests.get(url, headers=headers)
    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
    else:
        print(f"Error: {response.status_code}")
    return soup
```

```python
#Transform HTML, find all divs with class base-card relative w-full hover  

def transform(soup):
    
    divs = soup.find_all('div', class_="base-serp-page__content")
    return len(divs)
```
